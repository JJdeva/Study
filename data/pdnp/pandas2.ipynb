## set_index()

set_index로 특정 컬럼을 index를 지정하고, 다시 다른 컬럼을 set_index로 지정하면 기존의 인덱스는 삭제된다.

```python
set_df = df.copy()
```

```python
set_df = set_df.set_index('이름')
set_df.head(2)
```

```python
set_df = set_df.set_index('국어')
set_df.head(3) # 기존 이름컬럼을 인덱스로 지정한건 삭제된다.
```

## 시리즈 연산

- 시리즈 간의 연산
- 시리즈와 NaN

```python
student1 = pd.Series({'국어': 100, '영어': 90, '수학':95})
student2 = pd.Series({'국어': 90, '영어': 90, '수학':85})
```

```python
addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2
```

```python
addition, subtraction, multiplication, division
```

```python
result = pd.DataFrame([addition, subtraction, multiplication, division])
result
```

## NaN처리

```python
student1 = pd.Series({'국어': 100, '영어': 90, '수학':95})
student2 = pd.Series({'국어': 90, '수학':85})
```

```python
addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2
result = pd.DataFrame([addition, subtraction, multiplication, division])
result
```

함수 내 fill_value=0를 넣어주자

```python
addition = student1.add(student2, fill_value=0)
subtraction = student1.sub(student2, fill_value=0)
multiplication = student1.multiply(student2, fill_value=0)
division = student1.divide(student2, fill_value=0)
result = pd.DataFrame([addition, subtraction, multiplication, division])
result
```
