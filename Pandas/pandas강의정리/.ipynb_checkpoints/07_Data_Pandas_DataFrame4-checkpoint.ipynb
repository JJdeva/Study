# DataFrame에 새 column 추가 & 삭제

```python
import pandas as pd
```

```python
train_data = pd.read_csv('./train.csv')
train_data.head()
```

### 새 column 추가하기
- [] 사용하여 추가하기
- insert 함수 사용하여 원하는 위치에 추가하기

```python
train_data['Age_double'] = train_data['Age'] * 2
train_data.head() # 끝에 Age_double이라는 칼럼이 생김
```

```python
train_data['Age_triple'] = train_data['Age_double'] + train_data['Age']
train_data.head()
```

```python
train_data.insert(3, 'Fare10', train_data['Fare']/10) # 인덱스 3번에 Fare10을 만들고 그 값은 Fare를 10으로 나눈것이다
train_data.head() # 특정위치에 넣을 수 있는 것 = insert
```

### column 삭제하기
- drop 함수 사용하여 삭제
    - 리스트를 사용하여 멀티플 삭제 가능

```python
train_data.drop('Age_triple', axis=1).head()
```

```python
train_data.head() # drop을 해도 원본데이터에는 영향을 주지 않는다.
```

```python
train_data.drop(['Age_double', 'Age_triple'], axis=1).head()
```

```python
train_data.head()
```

```python
# 원본 자체에 삭제를 하고 싶을 때 ---> inplace를 조작해라
train_data.drop(['Age_double', 'Age_triple'], axis=1, inplace=True)
```

```python
train_data.head() # 삭제되었다.
```

# DataFrame Nan 데이터 처리

### NaN 값 확인
- info 함수를 통하여 개수 확인
- isna 함수를 통해 boolean 타입으로 확인

```python
train_data.info()
```

```python
train_data.isna() # True인 경우가 NaN
```

```python
train_data['Cabin'].isna()
```

### NaN 처리 방법
- 데이터에서 삭제
    - dropna 함수
- 다른 값으로 치환
    - fillna 함수

- NaN 데이터 삭제하기

```python
train_data.dropna() # NaN이 하나라도 있으면 행자체를 없애네
```

```python
train_data.dropna(subset=['Age']) # Age에 NaN이 있는 행만 지운다. subset 활용
```

```python
train_data.dropna(axis=1) # NaN이 있는 열을 지움. 기본은 axis=0
```

### NaN 값 대체하기
- 평균으로 대체하기
- 생존자/사망자 별 평균으로 대체하기

```python
train_data['Age'].mean()
```

```python
train_data['Age'].fillna(train_data['Age'].mean()) #888번째가 NaN이였는데 평균으로 대체되었다
```

```python
# 생존자 나이 평균
mean1 = train_data[train_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean0 = train_data[train_data['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

```python
train_data.loc[train_data['Survived']==1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data.loc[train_data['Survived']==0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```
