# stack, unstack 함수의 이해 및 활용

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })

df
```

### stack & unstack
- stack : 칼럼 레벨에서 인덱스 레벨로 dataframe 변경
    - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
- unstack : 인덱스 레벨에서 칼럼 레벨로 dataframe 변경
    - stack의 반대 operation
- 둘은 역의 관계에 있음

```python
new_df = df.set_index(['지역', '요일'])
new_df
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

# Concat 함수로 데이터 프레임 병합하기

### concat 함수 사용하여 dataframe 병합하기
- pandas.concat 함수
- 축을 따라 dataframe을 병합 가능
    - 기본 axis = 0 -> 행단위 병합

```python
from pandas import DataFrame as DF
```

```python
df1 = DF({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
df2 = DF({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df1
```

```python
pd.concat([df1, df2]) # axis=0 행으로 이어 붙임
```

```python
pd.concat([df1, df2], axis=1) # 열로 붙임
```

```python
df3 = DF({'key2' : np.arange(10), 'value2' : np.random.randn(10)})
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], axis=1)
```

# Merge & join 함수로 데이터 프레임 병합

### dataframe merge
- SQL의 join처럼 특정한 column을 기준으로 병합
    - join 방식 : how 파라미터를 통해 명시
        - inner : 기본값, 일치하는 값이 있을 경우
        - left : left outer join
        - right : right outer join
        - outer : full outer join
- pandas.merge 함수가 사용됨

```python
customer = DF({'customer_id' : np.arange(6),
              'name' : ['철수','영희','길동','영수','수민','동건'],
              '나이' : [40, 20, 21, 30, 31, 18]})
```

```python
customer
```

```python
orders = DF({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
            'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'],
            'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
orders.head()
```

- on
    - join 대상이 되는 column 명시

```python
pd.merge(customer, orders, on='customer_id', how='inner') # 두 테이블 비교해서 일치하는것만 가져옴
# 동건이는 물건이 없기에 누락, 9번손님은 0~5번 까지 있는 손님목록에 없어서 누락
```

```python
pd.merge(customer, orders, on='customer_id', how='left') # left : customer기준
```

```python
pd.merge(customer, orders, on='customer_id', how='right') # right : orders 기준
```

```python
pd.merge(customer, orders, on='customer_id', how='outer') # 둘이 합친거
```

- index 기준으로 join 하기

```python
cust1 = customer.set_index('customer_id')
order1 = orders.set_index('customer_id')
```

```python
cust1
```

```python
order1
```

```python
pd.merge(cust1, order1, left_index=True, right_index=True)
```

### 연습문제
1. 가장 많이 팔린 아이템은?
2. 영희가 가장 많이 구매한 아이템은?

```python
pd.merge(customer, orders, on='customer_id').groupby('item').sum().sort_values(by='quantity', ascending=False)
# quantity 기준으로 분류해주고, ascending=False == 내림차순으로 한다
```

```python
pd.merge(customer, orders, on='customer_id').groupby(['name','item']).sum().loc['영희', 'quantity']
```

```python
pd.merge(customer, orders, on='customer_id').groupby(['name','item']).sum()
```

### join 함수
- 내부적으로 pandas.merge 함수 사용
- 기본적으로 index를 사용하여 left join

```python
cust1.join(order1, how='inner')
```
